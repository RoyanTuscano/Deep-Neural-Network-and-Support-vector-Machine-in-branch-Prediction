```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.linear_model import SGDClassifier, Perceptron
```

```python
data = np.loadtxt('./ML_Data/CLIENT04_ML.csv', delimiter=',')
```

```python
samples=2000000
X = data[:samples,1:-2]
Y = data[:samples,-2:-1]
PC_Address=data[:samples,0]
```

```python
print(PC_Address.shape,Y.shape)
```

```python
batch_size=1
#Global Histroy with size of register 100
GHR=np.zeros((10,100))
#Local History with 10000 enteries of Local History parsed with the address. Each register size if 16
LHR=np.zeros((10000,16))
clf = Perceptron(penalty='l2')
X_input=np.concatenate((X[0:10],GHR[0:10],LHR[0:10]),axis=1)
clf.fit(X_input,Y[0:10])
print(X_input.shape,Y[0:10].shape,GHR[0,:].shape)
acc=0
add=int(PC_Address[0]%10000)
result=np.zeros((1,samples))
for i in range(0,samples,batch_size):
    add=int(PC_Address[i]%10000)
    x_input=np.concatenate((X[i:i+1],GHR[0:1],LHR[add:add+1]),axis=1)
    y_output=Y[i:i+1]
    y_pred=clf.predict(x_input)
    clf.partial_fit(x_input,y_output)
    GHR[0,:-1]=GHR[0,1:]
    GHR[0,-1]=Y[i:i+1,0]
    LHR[add,:-1]=LHR[add,1:]
    LHR[add,-1]=Y[i:i+1,0]
    # check if the output is correct and increment if correct
    if(y_pred==Y[i:i+1,0]):
        acc=acc+1
    # store the number of correct prediction at that particular point
    result[0,i]=acc
        
    
```

```python
temp=0
print(len(result[0]))
for i in range(int(len(result[0])/10000)):
    print(i,"cycle accuracy",(result[0,i*10000]-temp)/10000)
    temp=result[0,i*10000]
```
